```python
def get_start_index():
    index_page = 'https://www.ptt.cc/bbs/baseball/index.html'
    response = requests.get(index_page)
    soup = BeautifulSoup(response.text, 'html.parser')
    return int(soup.find_all('a', 'btn wide')[1]['href'].split('index')[1].replace('.html', ''))+1

def get_popularity(article):
    push = article.find('div', 'nrec').text
    if push == '':
        return 0
    elif push == '爆':
        return 100
    elif push[:1] == 'X':
        if push == 'XX':
            return -100
        else:
            return -10*int(push[1])
    return int(push)

def get_title(article):
    if '刪除) <' in article.find('div', 'title').text.strip():
        return '[deleted] deleted'
    return article.find('div', 'title').text.strip()

def split_cate_reply(article):
    title = get_title(article).split(']')
    reply = False
    if 'Re:' in title[0]:
        reply = True
    return title[0].split('[')[1].lower(), reply

def get_href(article):
    return article.a['href']
```

```python
def get_article(all_content:list) -> str:
    return ''.join([i for i in all_content[0].split('\n')[1:] if i != ''])

def get_comment_info(all_content:list) -> list:
    comments = all_content[1].split('\n')[1:-1]
    last_commenter = None
    comment_data = []
    for comment in comments:
        id_and_text = comment.split(': ') 
        try:    
            push = id_and_text[0].split()[0]
            id = id_and_text[0].split()[-1]
            text = re.sub(r'\d{2}/\d{2} \d{2}:\d{2}', '', id_and_text[1]).strip()
            if id == last_commenter:
                comment_data[-1][2] += text 
            elif id == '編輯':
                pass
            else:
                comment_data.append([push, id, text])   
        except IndexError:
            pass
        last_commenter = id
    return comment_data

response = requests.get(art_test)
soup = BeautifulSoup(response.text, 'html.parser')
main_content = soup.find('div', {'id': 'main-content'})
span_f2 = main_content.find('span', {'class': 'f2'})
all_content = main_content.get_text().split(span_f2.get_text())
article = get_article(all_content)
comment_info = get_comment_info(all_content)

```

```python
import sqlite3

# 建立一個資料庫連接
conn = sqlite3.connect('data.db')

# 建立一個游標物件
c = conn.cursor()

# 建立一個表
c.execute('''CREATE TABLE Article
             (
          board text, 
          popularity integer, 
          category text, 
          reply blob, 
          title text,
          article text,
          href text,
          article_id text,
          date text);
          ''')

c.execute('''CREATE TABLE Comment
             (
          article_id text,
          reaction text,
          author text,
          comment text);
          ''')

# # 插入一行資料
# c.execute("INSERT INTO stocks VALUES ('2006-01-05','BUY','RHAT',100,35.14)")

# 儲存 (commit) 變更
conn.commit()

# 關閉連接
conn.close()
```
